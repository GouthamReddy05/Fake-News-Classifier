```python
import pandas as pd

news_df1 = pd.read_csv('/content/news_dataset/Fake.csv')
news_df1.head()
```

```python
news_df2 = pd.read_csv('/content/news_dataset/True.csv')
news_df2.head()
```

```python
news_df1['label'] = 1
news_df2['label'] = 0
```

```python
news_df1.head()
```

```python
data = pd.concat([news_df1, news_df2], ignore_index=True)
data.head()
```

```python
import shutil

data = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
data = data.drop_duplicates().reset_index(drop=True)
```

```python
data = data.drop(['text','subject', 'date'], axis=1)
```

```python
x = data['title']
y = data['label']
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
```

```python
for i in range(0, len(x)):
  sen = re.sub('[^a-zA-Z]', ' ', x[i])
  sen = sen.lower()
  sen = sen.split()

  sen = [ps.stem(word) for word in sen if word not in stopwords.words('english')]
  sen = ' '.join(sen)
  corpus.append(sen)
```

```python
corpus
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=100000)
tokenizer.fit_on_texts(corpus)

sequences = tokenizer.texts_to_sequences(corpus)
```

```python
sequences
```

```python
padded_sequences = pad_sequences(sequences, padding='pre')
```

```python
padded_sequences
```

```python
len(padded_sequences[1])
```

```python
padded_sequences[0]
```

## Using LSTM



```python
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

sen_len = 35

model = Sequential()
model.add(Embedding(100000, 100, input_length=sen_len))

model.add(Dropout(0.2))

model.add(LSTM(400, return_sequences=True))
model.add(LSTM(400))

model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.build(input_shape=(None, 100))
model.summary()
```

```python
len(padded_sequences), y.shape
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Create EarlyStopping callback
early_stop = EarlyStopping(
    monitor='val_loss',     # or 'val_accuracy'
    patience=3,             # Number of epochs with no improvement before stopping
    restore_best_weights=True  # Restore weights from the best epoch
)
```

```python
import numpy as np
x = np.array(padded_sequences)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=64, callbacks=[early_stop])
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.where(y_pred > 0.6, 1, 0)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(f"Accuracy: {(accuracy_score(y_test, y_pred) * 100):.2f}")
```

```python
print(f"Classification Report: {classification_report(y_test, y_pred)}")
```

## Using GRU


```python
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import GRU, Bidirectional
sen_len = 34

model = Sequential()
model.add(Embedding(100000, 100, input_length=sen_len))

model.add(Dropout(0.2))

model.add(Bidirectional(GRU(200, return_sequences=True)))
model.add(Bidirectional(GRU(200)))

model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.build(input_shape=(None, 100))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
